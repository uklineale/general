```python
import nltk
from nltk.corpus import genesis, stopwords
import random as rand
```

```python
text = genesis.words('english-kjv.txt')
```

```python
stopList = stopwords.words('english')
```

```python
#http://stackoverflow.com/questions/1011938/python-previous-and-next-values-inside-a-loop
#Worth saving, allows you to get previous, current, and next values
from itertools import tee, islice, chain
try:
    #Python 2
    from itertools import izip
except:
    izip = zip

def previous_and_next(some_iterable):
    prevs, items, nexts = tee(some_iterable, 3)
    prevs = chain([None], prevs)
    nexts = chain(islice(nexts, 1, None), [None])
    return izip(prevs, items, nexts)
```

```python
#Shifts punctuation to it's proper place
#punctuationShift(['This', 'should', 'be', 'the', 'end', '.']) will output as 'This should be the end.'
#Does not deal with punctuation at the beginning of the sentence
test = ['I', 'am', 'born', 'again','.','This', 'should', 'be', 'the', 'end', '.', 'My', 'strength', 'is', 'futile', '.']

def punctuationShift(textList):
    punctuationList = '.?,;!'
    listSplit = previous_and_next(textList)
    built = ''
    for prevs, currs, nexts in listSplit:
        if nexts:
            if nexts not in punctuationList:
                if currs not in punctuationList:
                    built += currs + ' '
            else:
                built += currs + nexts + ' '
        else:
            if currs not in punctuationList:
                built += currs
    return built

punctuationShift(test)
```

```python
#Determines percentage of non-stopwords (content words)
def content_fraction(text, stoplist):
    content = [w for w in text if w.lower() not in stoplist]
    return len(content)/len(text)

content_fraction(text, stopList)
```

```python
#Access random word in O(1) rather than O(n): https://github.com/robtandy/randomdict
#Weight between stopword and non-stopword - weight proportionate to cfd
def generate(text, stoplist, length=50):
    #Generate cfd from text
    cfd = nltk.ConditionalFreqDist(nltk.bigrams(text))
    #List of content words - DOES NOT HAVE FREQUENCIES
    cleanCfdChoices = list(set(cfd.conditions()) - set(stopList))
    #List of all words - ALSO NO FREQS
    allCfdChoices = list(cfd.conditions())
    
    seed = rand.choice(cleanCfdChoices)
    length -= 1
    genString = [seed]
    
    for i in range(length):
        #get most common, pick random from most common, print, repeat
        seed = rand.choice(cfd[seed].most_common(3))[0]
        genString.append(seed)
    
    return punctuationShift(genString)
        
        
generate(text, stopList)
```
